```python
def legendre(a, p):
    """Kiểm tra a có là phần dư bậc hai mod p không"""
    return pow(a, (p - 1) // 2, p)

def tonelli_shanks(a, p):
    """Tìm căn bậc hai của a mod p, nếu có"""
    if legendre(a, p) != 1:
        return (None, None)  # Không có nghiệm

    # Trường hợp dễ: p ≡ 3 mod 4
    if p % 4 == 3:
        r = pow(a, (p + 1)//4, p)
        return (r, p - r)

    # Tách p-1 = q * 2^s
    q, s = p - 1, 0
    while q % 2 == 0:
        q //= 2
        s += 1

    # Tìm z là non-residue
    z = 2
    while legendre(z, p) != p - 1:
        z += 1

    # Khởi tạo
    m = s
    c = pow(z, q, p)
    t = pow(a, q, p)
    r = pow(a, (q + 1)//2, p)

    # Lặp hiệu chỉnh
    while t != 1:
        i = 1
        t2i = pow(t, 2, p)
        while t2i != 1:
            i += 1
            t2i = pow(t2i, 2, p)

        b = pow(c, 1 << (m - i - 1), p)
        r = (r * b) % p
        t = (t * b * b) % p
        c = (b * b) % p
        m = i
    return (r, p - r)

res = p
for i in ints:
    a, b = tonelli_shanks(i, p)
    if a != None:
        res = max(a, b)
print(res)    
```
