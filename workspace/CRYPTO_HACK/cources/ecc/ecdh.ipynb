```sage
from sage.all import *

p = 9739
# y2 +a1xy + a3y = x^3 + a2x^2 + a4x + a5
E = EllipticCurve(GF(p), [0, 0, 0, 497, 1768]) 
```

```sage
# =================== Curves and Logs =========================
import hashlib

G = E(1804,5368)

QA = E(815,3190)
nB = 1829

# TÃ­nh Key chung:
key = nB * QA 

flag = hashlib.sha1(str(key[0]).encode()).hexdigest()
print("crypto{" + flag + "}")
```
