```sage
from sage.all import *

p = 9739
# y2 +a1xy + a3y = x^3 + a2x^2 + a4x + a5
E = EllipticCurve(GF(p), [0, 0, 0, 497, 1768]) 
```

```sage
# ==================== Negative point ======================
P = E(8045, 6936)
Q = -P
print("crypto{" + str(Q[0]) + "," + str(Q[1]) + "}")
```

```sage
# ==================== Point addition ======================
P = E(493,5564)
Q = E(1539,4742)
R = E(4403,5202)

S = P + P + Q + R
print("crypto{" + str(S[0]) + "," + str(S[1]) + "}")
```

```sage
# ==================== Scalar Multiplication ======================
P = E(2339,2213)

Q = 7863 * P

print("crypto{" + str(Q[0]) + "," + str(Q[1]) + "}")
```
