```sage
from sage.all import *
from pwn import *
from Crypto.Util.number import isPrime, GCD

import fastecdsa
from fastecdsa.point import Point
from fastecdsa.curve import P256
from json import dumps

G = P256.G
p = P256.p
q = P256.q   # Order của đường cong ECC

Bx = 0x3B827FF5E8EA151E6E51F8D0ABF08D90F571914A595891F9998A5BD49DFA3531
By = 0xAB61705C502CA0F7AA127DEC096B2BBDC9BD3B4281808B3740C320810888592A
B = Point(Bx, By, curve=P256)

d = 5
while d != 1:
    if GCD(d, q) == 1:
        break
    d += 1


inv_d = pow(d, -1, q)

Binv = B * int(inv_d)

packet = {
    "private_key": int(d),
    "host": "www.bing.com",
    "curve": "secp256r1",
    "generator": [Binv.x, Binv.y]
}

host = "socket.cryptohack.org"
port = 13382

r = remote(host, port)

r.recvuntil('library!\n')
request = dumps(packet).encode()
r.sendline(request)

r.recvuntil('me about ').decode()
flag = r.recvline(keepends=False).strip().decode()

# r.interactive() 
r.close()

print(flag)
```
