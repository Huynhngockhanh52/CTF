**Gợi ý:** Bài toán này yêu cầu sử dụng các kỹ thuật trích xuất ảnh
1. `strings`: Là một công cụ hữu ích cho phép bạn trích xuất các chuỗi văn bản có thể đọc được từ tệp nhị phân, chẳng hạn như tệp hình ảnh. Khi chạy lệnh `strings` với tệp hình ảnh (ví dụ: `file.jpg`), nó sẽ tìm và hiển thị các chuỗi ký tự ***ASCII có độ dài tối thiểu là 4 ký tự*** trong tệp đó.

2. `steghide info`: Lệnh `steghide` trong Linux là một công cụ được sử dụng để ẩn thông tin trong các tệp hình ảnh và âm thanh. Khi sử dụng lệnh `steghide info` trên một tệp hình ảnh (như `file.jpg`), nó sẽ ***cung cấp thông tin về các dữ liệu đã được nhúng trong tệp đó***, nếu có.


Ta thực hiện trích xuất để lấy file được nhúng: `steghide extract -sf atbash.jpg`. 
 - `-sf`: Tham số này cho biết tệp nguồn (source file) mà bạn muốn trích xuất dữ liệu từ đó.
