```python
from Crypto.Util.number import getPrime, bytes_to_long
import random
import os
```

```python
e = 65537
p = getPrime(1024)
while (p - 1) % e == 0:
    p = getPrime(1024)
print(p)
```

```python
Ns = []
while len(Ns) < 4:
	q = getPrime(2048)
	N = p * q
	if (q - 1) % e != 0 and N not in Ns:
		Ns.append(N)
```

```python
import os
import random

# Thông điệp cần mã hóa
flag = 12345  # Ví dụ một số nguyên đại diện cho thông điệp

# Số mũ công khai
e = 65537  

# Danh sách các giá trị N (modulus)
Ns = [3233, 3919, 4957]  # Các modulus ví dụ (tích của hai số nguyên tố lớn)

# Bước 1: Mã hóa thông điệp với mỗi N trong danh sách Ns
cts = [pow(flag, e, N) for N in Ns]
print("Ciphertext (cts):", cts)

# Bước 2: Cập nhật danh sách Ns bằng cách thêm số ngẫu nhiên
Ns = [N + int(os.urandom(32).hex(), 16) for N in Ns]
print("Updated Ns:", Ns)

# Bước 3: Trộn ngẫu nhiên danh sách Ns
random.shuffle(Ns)
print("Shuffled Ns:", Ns)
```
