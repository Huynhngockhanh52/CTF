```python
from Crypto.Util import number

def gcd_extended(a, b):
    """Tính toán GCD và các hệ số x, y sao cho ax + by = gcd(a, b)"""
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = gcd_extended(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def mod_inverse(e, phi):
    """Tính toán nghịch đảo modulo của e với phi"""
    gcd, x, _ = gcd_extended(e, phi)
    if gcd != 1:
        raise Exception("Inverse does not exist")
    else:
        return x % phi

def rsa_decrypt(cipher_text, p, q, e):
    """Giải mã văn bản đã mã hóa bằng RSA"""
    # Bước 1: Tính N
    N = p * q
    print(N)
    # Bước 2: Tính phi(N)
    phi = (p - 1) * (q - 1)
    
    # Bước 3: Tính d
    d = mod_inverse(e, phi)
    print(d)
    # Bước 4: Giải mã
    plain_text = pow(cipher_text, d, N)
    
    return plain_text


```
