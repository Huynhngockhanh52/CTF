```python
from Crypto.Util.number import getPrime
from sympy import symbols, Poly, factor, sqrt, simplify
from sympy import Integer as symInt

p, q, r, s = [getPrime(512) for _ in "1234"]

# T√≠nh c√°c t·ªïng l≈©y th·ª´a ƒë·ªëi x·ª©ng
h1 = p + q + r + s
h2 = p**2 + q**2 + r**2 + s**2
h3 = p**3 + q**3 + r**3 + s**3
N = p * q * r * s

print("‚úÖ T·∫°o ng·∫´u nhi√™n th√†nh c√¥ng c√°c s·ªë nguy√™n t·ªë 512-bit.")
print("p =", p)
print("q =", q)
print("r =", r)
print("s =", s)

# T√≠nh l·∫°i c√°c h·ªá s·ªë ƒë·ªëi x·ª©ng a1, a2, a3, a4
a1 = h1
a4 = N

a2 = (a1**2 - h2) // 2
a3 = (h3 - a1**3 + 3*a1*a2) // 3

print("H·ªá s·ªë ƒëa th·ª©c ƒë·ªëi x·ª©ng:")
print(f"a1 = {a1}")
print(f"a2 = {a2}")
print(f"a3 = {a3}")
print(f"a4 = {a4}")

# X√¢y d·ª±ng l·∫°i ƒëa th·ª©c f(x) = x^4 - a1 x^3 + a2 x^2 - a3 x + a4
x = symbols('x')
f = x**4 - a1 * x**3 + a2 * x**2 - a3 * x + a4

print("Ph√¢n t√≠ch ƒëa th·ª©c...")
f_poly = Poly(f)
f_factors = factor(f_poly)

print("üìå ƒêa th·ª©c ph√¢n t√≠ch th√†nh:")
print(f_factors)

# Tr√≠ch nghi·ªám (c√°c s·ªë nguy√™n t·ªë p, q, r, s ban ƒë·∫ßu)
roots = f_poly.all_roots()
print("Nghi·ªám t√¨m ƒë∆∞·ª£c:")
for i, root in enumerate(roots, 1):
    print(f"root {i} = {root}")
```
